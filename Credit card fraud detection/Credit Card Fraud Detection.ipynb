## Pre-Processing

```python
from sklearn.preprocessing import StandardScaler
dataset['Normalised_Amount']=StandardScaler().fit_transform(dataset['Amount'].values.reshape(-1,1))
dataset=dataset.drop(columns=['Amount'])
```

```python
dataset.head()
```

```python
dataset=dataset.drop(['Time'],axis=1)
```

```python
dataset.head()
```

```python
X=dataset.drop(columns=['Class'])
y=pd.DataFrame(dataset.Class,columns=['Class'])
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
X_train=np.array(X_train)
X_test=np.array(X_test)
y_train=np.array(y_train)
y_test=np.array(y_test)
```

## Deep Neural Network

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
model=Sequential([
    Dense(units=16,input_dim=29,activation='relu'),
    Dense(units=24,activation='relu'),
    Dropout(0.5),
    Dense(20,activation='relu'),
    Dense(24,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

## Model Training

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x=X_train,
         y=y_train,
         epochs=5,
         batch_size=15)
```

```python
score=model.evaluate(x=X_test,
                    y=y_test)
```

```python
print(score)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(pd.DataFrame(y_test),y_pred.round())
```

```python
sns.heatmap(cm,annot=True,fmt='g')
```

```python
accuracy_score(pd.DataFrame(y_test),y_pred.round())*100
```
